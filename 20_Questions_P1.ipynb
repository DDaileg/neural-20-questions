# 20 Questions Game
Here is the first stage of the 20 questions game. The user selects a word from a predetermined list of words and answers yes/no questions that pertain to certain word attributes. The program then guesses the correct word based on the reported attributes. The words are stored in a csv file that also contains their attributes. In addition, there is a separate csv that contains the question for each attribute.

First let's import the libraries and files that I will be using. Since we are working with csv's, I will be using pandas to load the files as a dataframe

```python
import pandas as pd
```

```python
# Load data from csv files
df = pd.read_csv("data/word_attribute.csv")
qdf =pd.read_csv("data/attribute_question.csv")
```

Let's check to see if the data loaded properly

```python
df.head()
```

```python
qdf.head()
```

```python
# convert questions from questions csv to dictionary
questions = dict(zip(qdf["Attributes"], qdf["Questions"]))
```

```python
# prepare object and attribute list
objects = df["Name"].tolist()
attributes = df.columns[2:]
```

```python
# Start game
print('Think of one of these objects:')
print(', '.join(objects))
print("\nAnswer the following questions with 'y' or 'n'.") 
# initialize a dataframe that will be filtered by the given the answer choices
filtered_df = df.copy()
# Ask questions and filter through dataframe. If the program lands on the word early, the questions will stop
for attribute in attributes:
    question = questions.get(attribute)
    answer = input(f"{question} (y/n): ").strip().lower()
    if answer not in ["y", "n"]:
        print("Invalid input. Please answer with 'yes' or 'no'.")
        break

    expected = True if answer == "y" else False
    filtered_df = filtered_df[filtered_df[attribute] == expected]

    if len(filtered_df) == 1:
        break
        
# Provide the answer        
if len(filtered_df) == 1:
    guess = filtered_df["Name"].values[0]
    print(f"\nI guess you are thinking of: {guess}")
elif len(filtered_df) > 1:
    print("\nI'm not certain yet, but here are my best guesses:")
    print(", ".join(filtered_df["Name"].tolist()))
else:
    print("\nI couldn't guess your object. Maybe it's not in the database or the answers didn't match.")
```
